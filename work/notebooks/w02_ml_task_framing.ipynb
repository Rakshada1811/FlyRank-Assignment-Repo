This notebook frames the Content Refresh and Performance Analysis lane as a machine learning problem.

## 1. My lane as an ML task (type)

This problem is primarily a **classification** task. The objective is to classify whether a content page should be prioritized for review based on its observed performance metrics. Each page is assigned to a category such as requiring refresh or not requiring refresh using patterns learned from historical data.

```python
import pandas as pd

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

print('Dataset loaded successfully')
print('Rows:', len(df))
print('Columns:', len(df.columns))
```

## 2. Target or proxy

The prediction target is whether a page is likely to require a content refresh. Since a direct label is not available, the project uses the observed trend information provided in the dataset as a proxy. This allows the model to learn patterns associated with declining content performance.

```python
print('Trend direction counts:')
print(df['trend_direction'].value_counts())
```

## 3. Success metric

The primary evaluation metric is **Precision@50** because the goal is to recommend a small set of pages that are most likely to benefit from review. A higher Precision@50 means more of the highest-ranked recommendations are genuinely relevant.

```python
print('Evaluation Metric: Precision@50')
print('Higher values indicate better prioritization performance.')
```

## 4. The unit of analysis, as a real dataframe

Each row in the dataset represents **one content page** with multiple SEO-related features describing its performance.

```python
print(df.head())

print('\nShape:', df.shape)
```

## 5. Why ML beats a fixed rule here

Content performance depends on multiple interacting factors including impressions, clicks, CTR, search volume, position, and page characteristics. A simple rule based on a single threshold cannot effectively capture these relationships. Machine learning can combine many features simultaneously and identify more complex patterns to support better prioritization.

```python
numeric_columns = ['search_volume','impressions_90d','ctr','word_count']

print(df[numeric_columns].describe())
```
