# ML-04 — Search Intelligence Data Contract

This notebook defines the data contract for the Content Refresh and Performance Analysis lane.

## 1. Unit of analysis + time window

Each row represents **one content page** and its associated search performance metrics collected over the available reporting period. The dataset contains aggregated SEO features used to analyze whether a page may require content refresh.

```python
import pandas as pd

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

print('Rows:', len(df))
print('Columns:', len(df.columns))

print('\nSample data:')
print(df.head())
```

## 2. Fields: feature / label / context / excluded

**Features:** search_volume, impressions_90d, ctr, avg_position, word_count, content_age_days and other numerical SEO metrics.

**Label (Proxy):** trend_direction.

**Context:** content_type and other descriptive attributes that help interpret results.

**Excluded:** Any identifiers or fields that would not be available when making future predictions, or fields that could introduce information leakage. These fields are excluded because they do not help generalize the model.

```python
print('Dataset columns:')
for col in df.columns:
    print('-', col)
```

## 3. Verify it with queries (grain, counts, missing values, windows)

The following checks verify the dataset structure, missing values and basic statistics.

```python
print('Dataset Shape:', df.shape)

print('\nMissing Values:')
print(df.isnull().sum())

print('\nDuplicate Rows:', df.duplicated().sum())

print('\nSummary Statistics:')
print(df.describe(include='all'))
```

## 4. Data limits

This dataset provides observed search performance for anonymized pages. It supports identifying patterns that may help prioritize content review.

However, it cannot explain why rankings change, prove that refreshing content will improve rankings, or represent every factor considered by search engines. The results should therefore be interpreted as decision-support rather than causal evidence.

```python
print('Unique content types:', df['content_type'].nunique())

print('\nTrend distribution:')
print(df['trend_direction'].value_counts())
```
